```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#first i read the TotalData.csv into a dataframe
data=pd.read_csv('TotalData.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape # Prints the dimensions of the data frame-> rows: 5875335, columns:3
```

```python
# I use the dropna() function that drops all the lines that have nan values in them.
# In case there are nan values in Pressure and or time they will be droped as well
data=data.dropna()
```

```python
data.shape # now there are less lines because all nan values are gone
```

```python
data.head() # it seems that the first 52824 lines had nan values in them
```

# Start plotting

```python
# We first plot the histograms
data["Pressure"].plot(kind="hist",bins=1000,xlim=[1005,1035])
```

```python
data["Temperature"].plot(kind="hist",bins=100,xlim=[-5,30])
```

```python
data["frequency"].plot(kind="hist",bins=100,xlim=[0,60])
```

You can put these three histograms in the same plot

```python
fig, axs = plt.subplots(3, 1,figsize=(10,30))
data["Pressure"].plot(kind="hist",bins=1000,xlim=[1005,1035],ax=axs[0])
data["Temperature"].plot(kind="hist",bins=100,xlim=[-5,30],ax=axs[1])
data["frequency"].plot(kind="hist",bins=100,xlim=[0,60],ax=axs[2])
```

# Here i will create the scatter plots

```python
data.plot(x="Pressure",y="frequency",kind="scatter")
```

```python
# we see that there are a lot of values with pressure zero and this should be a mistake
# we can except them by asking for positive pressure
data.loc[data["Pressure"]>0.0].plot(x="Pressure",y="frequency",kind="scatter")
```

```python
# there are lots of data so the plots don't show well you can select values with a step of 1000
data[::1000].loc[data["Pressure"]>0.0].plot(x="Pressure",y="frequency",kind="scatter")
```
